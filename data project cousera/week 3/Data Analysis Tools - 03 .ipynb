```python

import pandas as pd
import scipy
import seaborn as sns
import matplotlib.pyplot as plt

data3 = pd.read_csv('./gapminder.csv', low_memory=False)
```

Set variable with numeric and discard NA values

```python
data3['internetuserate'] = pd.to_numeric(data3['internetuserate'], errors='coerce')
data3['incomeperperson'] = pd.to_numeric(data3['incomeperperson'], errors='coerce')
 
sub1_change = data3.copy()
 
data_clean_change = sub1_change.dropna()
```

Association between incomeperperson with internetuserate

```python
scipy.stats.pearsonr(data_clean_change['incomeperperson'], data_clean_change['internetuserate'])
```

```python
plt.figure()
scat1_change = sns.regplot(x="internetuserate", y="incomeperperson", fit_reg=True, data=data_clean_change)
plt.xlabel('Internet Use Rate')
plt.ylabel('Income Per Person')
plt.title('Scatterplot Between Internet Use Rate and Income Per Person')
plt.show()
```

The correlation between the internet use rate and GDP is remarkably strong, with a correlation coefficient of 0.814 and an extremely low p-value of 8.59 x 10-37. It is highly improbable for such a substantial relationship to occur by chance alone. This correlation coefficient suggests that approximately 66% of the variation in the internet use rate of a country can be predicted by knowing the GDP of that country.
